```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("my_csv_test.csv",header=None,)
```

```python
df.columns = ["num_heads","ff_dim","epochs","embed_dim","dropout_rate", 'mean', 'std']
df
```

```python
param_ = ["num_heads","ff_dim","epochs","embed_dim","dropout_rate"]

fig, ax = plt.subplots(2,3,figsize=(14,8), squeeze=False)
ax = ax.ravel()
for i in range(5):
    ax[i].set_title('Distribution of mean accuracy with {}'.format(param_[i]))
    sns.violinplot(x=param_[i],y='mean',data=df,ax=ax[i])
fig.tight_layout(pad=1.5)
```

```python
df.sort_values(by='mean',ascending = False)
```

```python
df.sort_values(by='mean',ascending = False).iloc[0:10]
```
