```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
Vectorizer = tf.keras.layers.TextVectorization(output_mode='tf_idf')
```

```python
Vectorizer.adapt(np.array(df['text']))
```

```python
vocab = Vectorizer.get_vocabulary()
```

```python
num_tokens = len(vocab)
print(num_tokens)
```

```python
model = tf.keras.Sequential([Vectorizer])
```

```python
print(vocab)
```

```python
model.predict(['the'])
```
